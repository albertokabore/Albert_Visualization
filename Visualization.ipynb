```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Create noisy sine wave

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Apply rolling mean (smoothing)
window_size = 3  # Define the window size for smoothing
df['y_smooth'] = df['y'].rolling(window=window_size).mean()  # Calculate rolling mean

# Plot original noisy data
plt.plot(df['x'], df['y'], label='Original Data', alpha=0.5)

# Plot smoothed data
plt.plot(df['x'], df['y_smooth'], label='Smoothed Data', color='blue')

# Add a legend to differentiate the two plots
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(-10, 10, 100)  # Generate 100 points between -10 and 10
y = x**2 + np.random.normal(0, 10, 100)  # Create noisy parabola (y = x^2 + noise)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Apply rolling mean (smoothing)
window_size = 5  # Define the window size for smoothing
df['y_smooth'] = df['y'].rolling(window=window_size).mean()  # Calculate rolling mean

# Plot original noisy data
plt.plot(df['x'], df['y'], label='Original', alpha=0.5)

# Plot smoothed data (rolling mean)
plt.plot(df['x'], df['y_smooth'], label='Smoothed Parabola', color='green')

# Add a legend to differentiate the two plots
plt.legend()

# Display the plot
plt.show()
```

```python
cars = data.cars()
cars.head()
```

The dataset contains the following columns:

Name: The name of the car.
Miles_per_Gallon: Fuel efficiency in miles per gallon.
Cylinders: Number of engine cylinders.
Displacement: Engine displacement in cubic inches.
Horsepower: Engine horsepower.
Weight_in_lbs: Weight of the car in pounds.
Acceleration: Time to accelerate from 0 to 60 mph (in seconds).
Year: Year of the car's model.
Origin: Country of origin (USA, Europe, or Japan).



```python
import altair as alt
from vega_datasets import data

# Load the cars dataset
cars = data.cars()

# Create a scatter plot
scatter_plot = alt.Chart(cars).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Horsepower', 'Miles_per_Gallon', 'Origin']
).properties(
    title='Horsepower vs. Miles per Gallon'
)

scatter_plot
```

```python
# Create a bar chart
bar_chart = alt.Chart(cars).mark_bar().encode(
    x='count()',
    y = 'Origin',
    color = 'Origin'
).interactive()

bar_chart
```

```python
# Create a histogram
histogram = alt.Chart(cars).mark_bar().encode(
    alt.X('Horsepower', bin=alt.Bin(maxbins=30)),
    y='count()',
    tooltip=['Horsepower', 'count()']
).properties(
    title='Distribution of Horsepower'
)

histogram
```

```python
# Create a line chart
line_chart = alt.Chart(cars).mark_line().encode(
    x='Year:T',
    y='average(Weight_in_lbs)',
    color='Origin'
).properties(
    title='Average Weight of Cars Over Time'
)

line_chart
```

```python
from plotnine import geom_boxplot

# Box plot for city MPG by number of cylinders
box_plot = (ggplot(mpg, aes(x='factor(cyl)', y='cty', fill='factor(cyl)')) +
            geom_boxplot() +
            ggtitle('City MPG by Number of Cylinders') +
            xlab('Number of Cylinders') +
            ylab('City MPG'))

print(box_plot)
```

```python
from plotnine import facet_wrap

# Facet plot for engine displacement vs highway MPG by drive type
facet_plot = (ggplot(mpg, aes(x='displ', y='hwy')) +
              geom_point() +
              facet_wrap('~drv') +
              ggtitle('Displacement vs. Highway MPG by Drive Type') +
              xlab('Engine Displacement (Liters)') +
              ylab('Highway MPG'))

print(facet_plot)
```

```python
from plotnine import ggplot, aes, geom_histogram, ggtitle, xlab, ylab, theme, element_rect, element_line

hist = (ggplot(mpg, aes(x='displ')) +
        geom_histogram(bins=30, fill='yellow', color='black') +
        theme (
            panel_background=element_rect(fill='lightblue'),
            panel_grid_major=element_line(color='white')
        )+
        ggtitle('Histogram'))

print(hist)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Sample data
df = px.data.gapminder()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Country Population Dashboard", style={'text-align': 'center'}),

    # Dropdown for selecting a year
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in df['year'].unique()],
        value=2007,  # Default value
        clearable=False,
        style={'width': '50%', 'margin': 'auto'}
    ),

    # Graph for population
    dcc.Graph(id='population-graph'),
])

# Define callback to update graph based on dropdown input
@app.callback(
    Output('population-graph', 'figure'),
    [Input('year-dropdown', 'value')]
)
def update_graph(selected_year):
    filtered_df = df[df['year'] == selected_year]
    fig = px.bar(filtered_df, x='country', y='pop', title=f'Population in {selected_year}', labels={'pop': 'Population'})
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Initialize the app
app = dash.Dash(__name__)

# Load the Gapminder dataset
df = px.data.gapminder()

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Gapminder Dashboard with Multiple Graphs", style={'textAlign': 'center'}),

    # Row 1: Scatter Plot and Bar Chart side by side
    html.Div([
        html.Div([
            dcc.Dropdown(
                id='scatter-continent-dropdown',
                options=[{'label': continent, 'value': continent} for continent in df['continent'].unique()],
                value='Asia',  # Default continent
                clearable=False,
                style={'width': '50%', 'margin': 'auto'}
            ),
            dcc.Graph(id='scatter-plot'),
        ], style={'width': '48%', 'display': 'inline-block'}),

        html.Div([
            dcc.Graph(id='bar-chart'),
        ], style={'width': '48%', 'display': 'inline-block'})
    ]),

    # Row 2: Line Chart spanning the entire width
    html.Div([
        dcc.Graph(id='line-chart')
    ], style={'width': '100%', 'display': 'inline-block'})
])

# Callback to update the scatter plot based on selected continent
@app.callback(
    dash.dependencies.Output('scatter-plot', 'figure'),
    [dash.dependencies.Input('scatter-continent-dropdown', 'value')]
)
def update_scatter(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    fig = px.scatter(filtered_df, x='gdpPercap', y='lifeExp', color='country',
                     title=f"Life Expectancy vs GDP per Capita in {selected_continent}",
                     log_x=True, size='pop', hover_name='country', size_max=60)
    return fig

# Callback to update the bar chart for population by continent
@app.callback(
    dash.dependencies.Output('bar-chart', 'figure'),
    [dash.dependencies.Input('scatter-continent-dropdown', 'value')]
)
def update_bar(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    pop_df = filtered_df.groupby('country').agg({'pop': 'sum'}).reset_index()
    fig = px.bar(pop_df, x='country', y='pop', title=f"Total Population by Country in {selected_continent}")
    return fig

# Callback to update the line chart for life expectancy over time
@app.callback(
    dash.dependencies.Output('line-chart', 'figure'),
    [dash.dependencies.Input('scatter-continent-dropdown', 'value')]
)
def update_line(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    fig = px.line(filtered_df, x='year', y='lifeExp', color='country',
                  title=f"Life Expectancy Over Time in {selected_continent}")
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
